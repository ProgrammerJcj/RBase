```r
# 검증하기
pred <- predict(model.rf, x)
# pred <- round(pred)

pred[pred <= 0.5] <-0
pred[pred > 0.5] <-1
```

```r
# 예측력과 오분류표
mean(pred == y)
table(pred, y)
```
