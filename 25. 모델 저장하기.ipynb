```r
library(randomForest)
```

```r
samp <- c(sample(1:50, 35), sample(51:100, 35), sample(101:150, 35))
iris.tr <- iris[samp,]
iris.test <- iris[-samp,]
iris.tr
```

```r
rf <- randomForest(Species ~., data=iris.tr, ntree=100)
```

```r
x <- subset(iris.test, select = -Species)
y <- iris.test$Species
```

```r
pred <- predict(rf, x)
```

```r
mean(pred == y)
table(pred, y)
```
