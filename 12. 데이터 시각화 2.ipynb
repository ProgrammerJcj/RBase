### 한화면에 차트 여러개 그려서 비교하기

```r
price <- runif(10,1,100)
plot(price)
```

```r
# 2행 2열 차트 그리기 : type 속성으로 구분해보기

par(mfrow = c(2,2))
plot(price, type='l')
plot(price, type='o')
plot(price, type='h')
plot(price, type='s')
```

```r
# 2행 2열 차트 그리기 : pch 속성으로 구분해보기

par(mfrow = c(2,2))

plot(price, type = 'o', pch = 5)
plot(price, type = 'o', pch = 15)
plot(price, type = 'o', pch = 20, col = 'blue')
plot(price, type = 'o', pch = c(rep(15,8),20,20), col = 'orange',
     cex = c(rep(1,8),3,1)
    )
```

```r
# 투표수가 높은 시군구의 후보별 투표수
```

```r
library(dplyr)
```

```r
election <- read.csv("../RFile/election_2017.csv",stringsAsFactors = F)
head(election)
```

```r
election_5 <-head(arrange(election,desc(election$pop)),5)
```

```r
시군구 <- as.matrix(t(election_5[,c(5:7)]))
```

```r
colnames(시군구) <- paste(election_5$광역시도,election_5$시군)
rownames(시군구) <-c("문재인","홍준표","안철수")
시군구
```

```r
par(family = "AppleGothic")
options(scipen = 100)

barplot(시군구,
        cex.names = 0.7,
        legend.text = T,
        ylim =c(0,500000),
        col = c("skyblue","red","lightgreen"),
        main = "투표수 높은 지역의 후보별 투표수"
       )

for(i in 1:5){
    text((0.7 + (1.2*(i-1))), sum(시군구[,i])+10000, sum(시군구[,i]), cex=0.8)
    text(0.7+((1.2)*(i-1)),시군구[1,i]/2,paste(round(시군구[1,i]/sum(시군구[,i])*100,2),"%"), cex=0.8)
    text(0.7+((1.2)*(i-1)),시군구[2,i]/2+시군구[1,i],paste(round(시군구[2,i]/sum(시군구[,i])*100,2),"%"), cex=0.8)
    text(0.7+((1.2)*(i-1)),시군구[3,i]/2+sum(시군구[1:2,i]),paste(round(시군구[3,i]/sum(시군구[,i])*100,2),"%"), cex=0.8)
}

```

# 수학 함수의 시각화
y = x^2

```r
f <- function(x){
    y=x^2
    return(y)
}
```

```r
x <- 1:3
f(x)
```

```r
x<-seq(-5,5,0.1)

plot(x,f(x),
     type="l"
    )
```
