```r
### 서울시 교통사고 데이터로 발생건수와 부상자수의 산포도 구하기
seoul <- read.csv("../RFile/newSeoul2019.csv")
head(seoul)
```

```r
# 발생건수의 숫자가 150이상이면 빨간색, 아니면 파란색으로 표현하기
par(family = "AppleGothic")

plot(seoul$발생건수, seoul$부상자수,
     main = "서울시 교통사고 발생건수와 부상자수 산포도",
     xlab = "발생건수",
     ylab = "부상자수",
     col = ifelse(seoul$발생건수 >= 150,'red','blue')
    )

plot(x, y,
     xlim = c(11,0),
     ylim = c(0,11)
    )
```

```r
#main title = Simple series
y <- c(1,3,2,4,8,5,7,6,9)
```

```r
# 꺽은 선 그래프
par(family = "AppleGothic")

plot(y,
     main = "Simple series",
     xlim = c(min(y)-1,length(y)+1),
     ylim = c(min(y)-1,max(y)+1),
     type = "b"

    )
```

```r
# 꺽은선 그래프

par(family = "AppleGothic")

plot(y,
     main = "Simple series",
     xlim = c(min(y)-1,length(y)+1),
     ylim = c(min(y)-1, max(y)+1),
     type = "c"

    )
```

```r
# 점이 빠진 꺾은 선 그래프

y <- c(1,3,2,4,8,5,7,6,9)

par(family = "AppleGothic")

plot(y,
     main = "Simple series",
     xlim = c(min(y)-1,length(y)+1),
     ylim = c(min(y)-1, max(y)+1),
     type = "o",
     col = c(rep("red",3),"blue", rep("green",6))

    )
```

```r
# 수직선 그래프

y <- c(1,3,2,4,8,5,7,6,9)

par(family = "AppleGothic")

plot(y,
     main = "Simple series",
     xlim = c(min(y)-1,length(y)+1),
     ylim = c(min(y)-1, max(y)+1),
     type = "h",
     col = c(rep("red",3),"blue", rep("green",6))

    )
```

```r
# 왼쪽 값을 기초로한 계단 모양 그래프


y <- c(1,3,2,4,8,5,7,6,9)

par(family = "AppleGothic")

plot(y,
     main = "Simple series",
     xlim = c(min(y)-1,length(y)+1),
     ylim = c(min(y)-1, max(y)+1),
     type = "s",

    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프


y <- c(1,3,2,4,8,5,7,6,9)

par(family = "AppleGothic")

plot(y,
     main = "Simple series",
     xlim = c(min(y)-1,length(y)+1),
     ylim = c(min(y)-1, max(y)+1),
     type = "S",

    )
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프


y <- c(1,3,2,4,8,5,7,6,9)

par(family = "AppleGothic")

plot(y,
     main = "Simple series",
     xlim = c(min(y)-1,length(y)+1),
     ylim = c(min(y)-1, max(y)+1),
     type = "s",
     lty = 3

    ) 
```

```r
# 오른쪽 값을 기초로한 계단 모양 그래프


y <- c(1,3,2,4,8,5,7,6,9)

par(family = "AppleGothic")

plot(y,
     main = "Simple series",
     xlim = c(min(y)-1,length(y)+1),
     ylim = c(min(y)-1, max(y)+1),
     pch = c(24,20,9),
     col = c("red","blue","yellow")

    ) 
```

```r
x <- runif(100)
y <- runif(100)

# pch : y > 0.5 이면 1, 아니면 18
# col : y > 0.5 이면 "red", 아니면 "blue"


par(family = "AppleGothic")

plot(x,y,
     main = "Simple series",
     xlim = c(min(y), max(y)),
     ylim = c(min(y), max(y)),
     pch = c(ifelse(y>0.5,1,18)),
     col = c(ifelse(y >0.5,"red","blue"))

    ) 
```
